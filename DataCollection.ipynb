```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
haar_data = cv2.CascadeClassifier('data.xml')
```

# Live Face Data Collection Without Mask 

```python
capture = cv2.VideoCapture(0)
WithoutMask_data = []
while True :
    flag, img = capture.read()
    if flag :
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces : 
            cv2.rectangle(img, (x,y), (x+w, y+h), (245,224,69), 2)
            face = img[y: y+h, x:x+w, :]
            face = cv2.resize(face, (50,50))
            print(len(WithoutMask_data))
            if len(WithoutMask_data) < 200 :
                WithoutMask_data.append(face)
        cv2.imshow("Capturing Without Mask Images ..!",img)
        if cv2.waitKey(2) == 27 or len(WithoutMask_data) >= 200 :
            break

capture.release()
cv2.destroyAllWindows()
```

```python
np.save("Without_Mask.npy",WithoutMask_data)
```

# Live Face Data Collection With Mask

```python
capture = cv2.VideoCapture(0)
WithMask_data = []
while True :
    flag, img = capture.read()
    if flag :
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces : 
            cv2.rectangle(img, (x,y), (x+w, y+h), (245,224,69), 2)
            face = img[y: y+h, x:x+w, :]
            face = cv2.resize(face, (50,50))
            print(len(WithMask_data))
            if len(WithMask_data) < 200 :
                WithMask_data.append(face)
        cv2.imshow("Capturing With Mask Images..!",img)
        if cv2.waitKey(2) == 27 or len(WithMask_data) >= 200 :
            break

capture.release()
cv2.destroyAllWindows()
```

```python
np.save("With_Mask.npy",WithMask_data)
```
